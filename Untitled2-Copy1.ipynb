## 1. ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Ï∂îÏ∂ú
<br>

Ï∂îÏ≤ú ÏãúÏä§ÌÖúÏùÄ **ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏãúÏä§ÌÖú**ÏùÑ ÏÇ¨Ïö©ÌïúÎã§. ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÄ Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ÏóêÏÑú ÏûÖÎ†• Ìè¨Ïù∏Ìä∏Ïóê Í∞ÄÏû• Í∞ÄÍπåÏö¥ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º Ï∞æÎäî Í≥ºÏ†ïÏùÑ ÎÇòÌÉÄÎÇ∏Îã§.

üëâ Îã§ÏñëÌïú ÌÅ¥Î†àÏä§Ïóê ÎåÄÌïú ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïùò Í∑ºÏ†ëÏÑ±ÏùÑ Í∏∞Î∞òÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º Î∂ÑÎ•òÌïòÎäî ÏãúÏä§ÌÖúÏùÑ Íµ¨Ï∂ïÌïòÎäîÎç∞ ÏûêÏ£º ÏÇ¨Ïö©



<br>
<br>
ÏÉà ÌååÏù¥Ïç¨ ÌååÏùº ÏÉùÏÑ±ÌïòÍ≥† Îã§Ïùå Ìå®ÌÇ§ÏßÄÎ•º ÏûÑÌè¨Ìä∏ÌïúÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

<br>
ÏÉòÌîå 2D Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º Ï†ïÏùòÌïúÎã§.

```python
# ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞
X = np.array([[2.1,1.3], [1.3, 3.2], [2.9, 2.5], [2.7, 5.4], [3.8, 0.9],
            [7.3, 2.1], [4.2, 6.5], [3.8, 3.7], [2.5, 4.1], [3.4, 1.9],
            [5.7, 3.5], [6.1, 4.3], [5.1, 2.2], [6.2, 1.1]])
```

<br>Ï∂îÏ∂úÌïòÎ†§Îäî ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏàòÎ•º Ï†ïÏùòÌïúÎã§.

```python
# ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Ïàò
k = 5
```

<br>K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ Ï∂îÏ∂úÌïòÎäî Îç∞ ÏÇ¨Ïö©Ìï† ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º Ï†ïÏùòÌïúÎã§.

```python
# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏
test_data_point = [4.3, 2.7]
```

<br>ÏõêÌòï Í≤ÄÏ†ïÏÉâ ÎßàÏª§Î•º ÏÇ¨Ïö©Ìï¥ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î¶∞Îã§.

```python
# ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î¶∞Îã§
plt.figure()
plt.title('Input data')
plt.scatter(X[:,0], X[:,1], marker='o', s=75, color='black')
```

- ##### ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ ÏãúÍ∞ÅÌôî

<br>
<br>
<br>


ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î™®Îç∏ÏùÑ ÏÉùÏÑ±ÌïòÍ≥† ÌõàÎ†®ÌïúÎã§. Î™®Îç∏ÏùÄ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïóê Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ïù¥ÏõÉÏùÑ Ï∂îÏ∂úÌïòÎäîÎç∞ ÏÇ¨Ïö©ÌïúÎã§.

```python
# K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î™®Îç∏ ÏÉùÏÑ±
knn_model = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(X)
distances, indices = knn_model.kneighbors([test_data_point])
```

<br>
Î™®Îç∏ÏóêÏÑú Ï∂îÏ∂úÌïú ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ Ï∂úÎ†•ÌïúÎã§.

üëâ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïóê Í∞ÄÏû• Í∞ÄÍπåÏö¥ Ìè¨Ïù∏Ìä∏ 5Í∞úÎ•º Î≥¥Ïó¨Ï§ÄÎã§.

```python
#ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ KÍ∞ú Ï∂úÎ†•
print("\nK Nearest Neighbors:")
for rank, index in enumerate(indices[0][:k], start=1):
    print(str(rank) + " ==>", X[index])
```

- ##### K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Í≤∞Í≥º

<br>
<br>
<br>
<br>ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ ÏãúÍ∞ÅÌôîÌïúÎã§.

```python
# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ÏôÄ ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ ÏãúÍ∞ÅÌôî
plt.figure()
plt.title('Nearest neighbors')
plt.scatter(X[:, 0], X[:, 1], marker='o', s=75, color='k')
plt.scatter(X[indices][0][:][:, 0], X[indices][0][:][:, 1], 
        marker='o', s=250, color='k', facecolors='none')
plt.scatter(test_data_point[0], test_data_point[1],
        marker='x', s=75, color='k')

plt.show()
```

- ##### ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ 5Í∞ú ÏãúÍ∞ÅÌôî

<br>

## 2. K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î∂ÑÎ•òÍ∏∞ ÏÉùÏÑ±
<br>

**K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î∂ÑÎ•òÍ∏∞** : Ï£ºÏñ¥ÏßÑ Îç∞Ïù¥ÌÑ∞Î•º Î∂ÑÎ•òÌïòÎäî Îç∞ K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÏÇ¨Ïö©ÌïòÎäî Î∂ÑÎ•ò Î™®Îç∏

1. ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏ÏóêÏÑú Í∞ÄÏû• Í∞ÄÍπåÏö¥ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ KÍ∞úÎ•º Ï∞æÎäîÎã§.
2. Í≥ºÎ∞òÏàò Ìà¨ÌëúÎ•º Í∏∞Î∞òÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïóê ÌÅ¥ÎûòÏä§Î•º Ìï†ÎãπÌïúÎã§.
3. Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ KÍ∞úÏùò Î™©Î°ùÏóêÏÑú ÌÅ¥ÎûòÏä§Î•º Î≥¥Í≥† ÌëúÎ•º Í∞ÄÏû• ÎßéÏù¥ ÏñªÏùÄ ÌÅ¥ÎûòÏä§Î•º ÏÑ†ÌÉùÌïúÎã§. 

üëâ K Í∞íÏùÄ Î¨∏Ï†úÏóê Îî∞Îùº Îã¨ÎùºÏßê


<br>
<br>
ÏÉà ÌååÏù¥Ïç¨ ÌååÏùºÏùÑ ÎßåÎì§Í≥† Îã§Ïùå Ìå®ÌÇ§ÏßÄÎ•º ÏûÑÌè¨Ìä∏ ÌïúÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import neighbors, datasets
```

<br>

ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Îäî data.txt ÌååÏùºÏóêÏÑú Î°úÎìúÌïúÎã§. Í∞Å ÎùºÏù∏ÏóêÏÑú Í∞íÏùÄ ÏâºÌëúÎ°ú Î∂ÑÎ¶¨ÎêòÎ©∞ Îç∞Ïù¥ÌÑ∞Îäî ÌÅ¥ÎûòÏä§ 4Í∞úÎ•º Ìè¨Ìï®ÌïúÎã§.

```python
# ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ Î°úÎìú
input_file = 'data.txt'
data = np.loadtxt(input_file, delimiter=',')
X, y = data[:, :-1], data[:, -1].astype(int)
```

<br>
ÎÑ§ Í∞ÄÏßÄ ÎßàÏª§Î•º ÏÇ¨Ïö©Ìï¥ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌïúÎã§. Î†àÏù¥Î∏îÏùÑ ÎßàÏª§Ïóê Îß§ÌïëÌï¥ mapper Î≥ÄÏàòÍ∞Ä Í∑∏Î¶ºÏóê ÌëúÏãúÎêòÎèÑÎ°ù ÌïúÎã§. 

```python
# ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Í∑∏Î¶∞Îã§.
plt.figure()
plt.title('Input data')
marker_shapes = 'v^os'
mapper = [marker_shapes[i] for i in y]
for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[i], 
            s=75, edgecolors='black', facecolors='none')
```

- ### ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî
<br>
<br>
<br>
ÏÇ¨Ïö©Ìï† ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏàòÎ•º Ï†ïÌïúÎã§.

```python
# ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Ïàò
num_neighbors = 12
```

<br>
Î∂ÑÎ•òÍ∏∞ Î™®Îç∏Ïùò Í≤ΩÍ≥ÑÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Îç∞ ÏÇ¨Ïö©Ìï† Í∑∏Î¶¨ÎìúÏùò Ïä§ÌÖù ÌÅ¨Í∏∞Î•º Ï†ïÏùòÌïúÎã§.

```python
# ÏãúÍ∞ÅÌôî Í∑∏Î¶¨ÎìúÏùò Ïä§ÌÖù ÌÅ¨Í∏∞
step_size = 0.01  
```

<br>
K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î∂ÑÎ•òÍ∏∞ Î™®Îç∏ÏùÑ ÏÉùÏÑ±ÌïúÎã§.

```python
# K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î∂ÑÎ•òÍ∏∞ Î™®Îç∏ ÏÉùÏÑ±
classifier = neighbors.KNeighborsClassifier(num_neighbors, weights='distance')
```

<br>
ÌõàÎ†® Îç∞Ïù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï¥ Î™®Îç∏ÏùÑ ÌõàÎ†®ÌïúÎã§.

```python
# K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Î™®Îç∏ ÌõàÎ†®
classifier.fit(X, y)
```

<br>
Í∑∏Î¶¨ÎìúÎ•º ÏãúÍ∞ÅÌôîÌïòÎäî Îç∞ ÏÇ¨Ïö©Ìï† Í∞íÎì§Ïùò Î©îÏãú Í∑∏Î¶¨ÎìúÎ•º ÏÉùÏÑ±ÌïúÎã§,

```python
# Í≤ΩÍ≥ÑÎ•º Í∑∏Î¶¨Í∏∞ ÏúÑÌïú Î©îÏãú ÏÉùÏÑ±
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
x_values, y_values = np.meshgrid(np.arange(x_min, x_max, step_size), 
        np.arange(y_min, y_max, step_size))
```

<br>
Í≤ΩÍ≥ÑÎ•º ÏãúÍ∞ÅÌôîÌïòÍ∏∞ ÏúÑÌï¥ Í∑∏Î¶¨ÎìúÏùò Î™®Îì† Ï†êÏóê ÎåÄÌï¥ Î∂ÑÎ•òÍ∏∞Î•º ÌèâÍ∞ÄÌïúÎã§.

```python
# Í∑∏Î¶¨ÎìúÏùò Î™®Îì† Ï†êÏóê ÎåÄÌï¥ Î∂ÑÎ•òÍ∏∞ ÌèâÍ∞Ä
output = classifier.predict(np.c_[x_values.ravel(), y_values.ravel()])
```

<br>
Í≤∞Í≥ºÎ•º ÏãúÍ∞ÅÌôîÌïòÍ∏∞ ÏúÑÌï¥ Ïª¨Îü¨ Î©îÏãúÎ•º ÏÉùÏÑ±ÌïúÎã§. Ïª¨Îü¨ Î©îÏãú ÏúÑÏóê ÌõàÎ†® Îç∞Ïù¥ÌÑ∞Î•º Ïò§Î≤ÑÎ†àÏù¥Ìï¥ Í≤ΩÍ≥ÑÎ•º Í∏∞Ï§ÄÏúºÎ°ú Îç∞Ïù¥ÌÑ∞Î•º ÏãúÍ∞ÅÌôîÌïúÎã§.

```python
for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[i], 
            s=75, edgecolors='black', facecolors='none')
    
# ÏòàÏ∏° Í≤∞Í≥º ÏãúÍ∞ÅÌôî
output = output.reshape(x_values.shape)
plt.title('K Nearest Neighbors classifier model boundaries')
plt.pcolormesh(x_values, y_values, output, cmap=cm.Paired)

# ÎßµÏóê ÌõàÎ†®Îêú Ìè¨Ïù∏Ìä∏ Ïò§Î≤ÑÎ†àÏù¥
for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[i], 
            s=50, edgecolors='black', facecolors='none')
```

- ### Î∂ÑÎ•òÍ∏∞ Î™®Îç∏ Í≤ΩÍ≥Ñ
<br>
<br>
<br>
Ï†úÎ™©Í≥º X, Y ÌïúÍ≥ÑÎ•º ÏÑ§Ï†ïÌïúÎã§.

```python
plt.xlim(x_values.min(), x_values.max())
plt.ylim(y_values.min(), y_values.max())
plt.title('K Nearest Neighbors classifier model boundaries')
```

<br>
<br>
<br>
Î∂ÑÎ•òÍ∏∞Í∞Ä Ïñ¥ÎñªÍ≤å ÏûëÎèôÌïòÎäîÏßÄ Î≥¥Í∏∞ ÏúÑÌï¥ Îç∞Ïä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Î•º Ï†ïÏùòÌïúÎã§. ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Í∞Ä ÏûàÎäî Í∑∏Î¶ºÏùÑ ÏÉùÏÑ±Ìï¥ Í∞ÅÍ∞ÅÏùò ÏúÑÏπòÎ•º Ï∂úÎ†•ÌïúÎã§.

```python
# ÌÖåÏä§Ìä∏ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏
test_datapoint = [5.1, 3.6]
plt.figure()
plt.title('Test datapoint')
for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[i], 
            s=75, edgecolors='black', facecolors='none')

plt.scatter(test_datapoint[0], test_datapoint[1], marker='x', 
        linewidth=6, s=200, facecolors='black')
```

- ### ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ ÏÑ∏Ìä∏Ïóê ÏÉÅÎåÄÏ†ÅÏù∏ ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ 
<br>
<br>
<br>
Î∂ÑÎ•òÍ∏∞ Î™®Îç∏ÏùÑ Í∏∞Î∞òÏúºÎ°ú ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏Ïùò K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ Ï∂îÏ∂úÌïúÎã§.

```python
# K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Ï∂îÏ∂ú
_, indices = classifier.kneighbors([test_datapoint])
indices = indices.astype(int)[0]
```

<br>
K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ Í∑∏Î¶∞Îã§. Í∑∏ ÌõÑÏóê ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Ìè¨Ïù∏Ìä∏ÏôÄ ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î•º Ïò§Î≤ÑÎ†àÏù¥ÌïúÎã§

```python
# K-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉÏùÑ Í∑∏Î¶∞Îã§
plt.figure()
plt.title('K Nearest Neighbors')

for i in indices:
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[y[i]], 
            linewidth=3, s=100, facecolors='black')

plt.scatter(test_datapoint[0], test_datapoint[1], marker='x', 
        linewidth=6, s=200, facecolors='black')

for i in range(X.shape[0]):
    plt.scatter(X[i, 0], X[i, 1], marker=mapper[i], 
            s=75, edgecolors='black', facecolors='none')
```

- ### 12Í∞úÏùò ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ Í∑∏ÎûòÌîÑ
<br>
<br>
<br>
ÏòàÏ∏° Í≤∞Í≥ºÎ•º Ï∂úÎ†•ÌïúÎã§.

```python
print("Predicted output:", classifier.predict([test_datapoint])[0])

plt.show()
```
